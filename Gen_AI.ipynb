```python
import pandas as pd
import string
df = pd.read_csv('/content/Twitter Sentiments.csv')
df = df.drop(columns=['id', 'label'])
print(df.head())
```

### Install and Import NLTK
First, we need to install the `nltk` library and download the necessary data for stop words, tokenization, and lemmatization/stemming.

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('punkt_tab') # Added to resolve LookupError

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer, PorterStemmer
```

### 1. Convert to Lowercase
Converting text to lowercase helps in treating words like 'Hello' and 'hello' as the same, reducing the vocabulary size and improving consistency.

```python
df['tweet_lower'] = df['tweet'].apply(lambda x: x.lower())
print("After lowercase conversion:")
print(df[['tweet', 'tweet_lower']].head())
```

### 2. Remove Punctuation
Removing punctuation marks helps clean the text, as punctuation usually doesn't contribute to the sentiment or meaning of words in many NLP tasks. We'll use the `string` module for this.

```python
import string
def remove_punctuation(text):
    return ''.join([char for char in text if char not in string.punctuation])

df['tweet_no_punct'] = df['tweet_lower'].apply(remove_punctuation)
print("\nAfter punctuation removal:")
print(df[['tweet_lower', 'tweet_no_punct']].head())
```

### 3. Stop Word Removal
Stop words are common words (like 'the', 'is', 'a') that often carry little significant meaning and can be removed to focus on more important words. We use NLTK's English stop words list.

```python
stop_words = set(stopwords.words('english'))

def remove_stopwords(text):
    words = word_tokenize(text)
    return ' '.join([word for word in words if word not in stop_words])

df['tweet_no_stopwords'] = df['tweet_no_punct'].apply(remove_stopwords)
print("\nAfter stop word removal:")
print(df[['tweet_no_punct', 'tweet_no_stopwords']].head())
```

### 4. Tokenization
Tokenization is the process of breaking down text into individual words or tokens. This is a fundamental step for most NLP tasks.

```python
df['tweet_tokens'] = df['tweet_no_stopwords'].apply(word_tokenize)
print("\nAfter tokenization:")
print(df[['tweet_no_stopwords', 'tweet_tokens']].head())
```

### 5. Lemmatization
Lemmatization reduces words to their base or dictionary form (lemma). For example, 'running', 'runs', 'ran' all become 'run'. This helps in reducing inflected forms of a word to a common base form.

```python
lemmatizer = WordNetLemmatizer()

def lemmatize_words(tokens):
    return [lemmatizer.lemmatize(word) for word in tokens]

df['tweet_lemmatized'] = df['tweet_tokens'].apply(lemmatize_words)
print("\nAfter lemmatization:")
print(df[['tweet_tokens', 'tweet_lemmatized']].head())
```

### 6. Stemming
Stemming is a process of reducing inflected (or sometimes derived) words to their word stem, base or root form窶波enerally a written word form. The stem is not necessarily identical to the morphological root of the word; it is usually an affix that we can attach different suffixes to.

```python
stemmer = PorterStemmer()

def stem_words(tokens):
    return [stemmer.stem(word) for word in tokens]

df['tweet_stemmed'] = df['tweet_tokens'].apply(stem_words)
print("\nAfter stemming:")
print(df[['tweet_tokens', 'tweet_stemmed']].head())
```

### 7. Sentence Tokenization
Sentence tokenization (or sentence segmentation) is the process of splitting written language into its component sentences. This is useful for tasks that require analysis at the sentence level.

```python
from nltk.tokenize import sent_tokenize

def tokenize_sentences(text):
    return sent_tokenize(text)

# We will apply sentence tokenization on the original 'tweet_no_punct' column
# since sentence structure might be lost after word tokenization or stemming.
df['tweet_sentences'] = df['tweet_no_punct'].apply(tokenize_sentences)
print("\nAfter sentence tokenization:")
print(df[['tweet_no_punct', 'tweet_sentences']].head())
```
